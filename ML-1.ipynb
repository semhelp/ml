```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# Select numerical column (e.g., petal length)
num_col = 'petal length (cm)'
mean = df[num_col].mean()
median = df[num_col].median()
mode = df[num_col].mode()[0]  # mode returns a Series, take the first value
std = df[num_col].std()
variance = df[num_col].var()
range_val = df[num_col].max() - df[num_col].min()

print(f"Mean: {mean}, Median: {median}, Mode: {mode}, Std: {std}, Variance: {variance}, Range: {range_val}")

# Histogram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(df[num_col], kde=True)
plt.title('Histogram of Petal Length')

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=df[num_col])
plt.title('Boxplot of Petal Length')
plt.tight_layout()
plt.show()

# Outliers using IQR
Q1 = df[num_col].quantile(0.25)
Q3 = df[num_col].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df[num_col] < (Q1 - 1.5 * IQR)) | (df[num_col] > (Q3 + 1.5 * IQR))][num_col]

# Categorical variable (species) frequency
cat_col = 'species'
freq = df[cat_col].value_counts()
plt.figure(figsize=(6, 4))
freq.plot(kind='pie', title='Species Frequency')
plt.show()
```
